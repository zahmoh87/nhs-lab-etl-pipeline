```python
import pandas as pd
# Load CSV file
df = pd.read_csv('/content/lab_results.csv')
```

```python
#remove duplicate
df = df.drop_duplicates()
# Fill missing values
df['Result'] = df['Result'].fillna('Unknown')
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Date'] = df['Date'].fillna(pd.Timestamp('2025-01-01'))
```

```python
# Add abnormal flag
df['IsAbnormal'] = df['Result'].apply(lambda x: 1 if x == 'Abnormal' else 0)

# Group by lab
summary = df.groupby('LabLocation')['IsAbnormal'].sum().reset_index()
print(summary)
```

```python
# Save to new CSV
summary.to_csv('abnormal_summary.csv', index=False)
```
